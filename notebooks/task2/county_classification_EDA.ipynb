```python
# prompt: stacked histogram to show the metro, micropolitant and non-metro county count distribution in states

import matplotlib.pyplot as plt

# Group data by state and metro/micro/non-metro status
metro_counts = sub_county_df.groupby('State')['Metro2013'].sum().reset_index()
micro_counts = sub_county_df.groupby('State')['Micropolitan2013'].sum().reset_index()
nonmetro_counts = sub_county_df.groupby('State')['Nonmetro2013'].sum().reset_index()

# Create the stacked histogram
plt.figure(figsize=(12, 6))
plt.bar(metro_counts['State'], metro_counts['Metro2013'], label='Metro')
plt.bar(micro_counts['State'], micro_counts['Micropolitan2013'], bottom=metro_counts['Metro2013'], label='Micropolitan')
plt.bar(micro_counts['State'], nonmetro_counts['Nonmetro2013'], bottom=metro_counts['Metro2013'] + micro_counts['Micropolitan2013'], label='Nonmetro')

plt.xlabel('State')
plt.ylabel('County Count')
plt.title('Distribution of Metro, Micro, and Non-Metro Counties by State')
plt.xticks(rotation=90)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
urban_influence_2003 = sub_county_df.groupby('State')['UrbanInfluenceCode2003'].value_counts().unstack(fill_value=0)
urban_influence_2013 = sub_county_df.groupby('State')['UrbanInfluenceCode2013'].value_counts().unstack(fill_value=0)

plt.figure(figsize=(12, 6))

bottom = np.zeros(len(urban_influence_2003))
for col in urban_influence_2003.columns:
    plt.bar(urban_influence_2003.index, urban_influence_2003[col], bottom=bottom, label=f'2003 - {col}')
    bottom += urban_influence_2003[col]

bottom = np.zeros(len(urban_influence_2013))
for col in urban_influence_2013.columns:
    plt.bar(urban_influence_2013.index, urban_influence_2013[col], bottom=bottom, label=f'2013 - {col}')
    bottom += urban_influence_2013[col]


plt.xlabel('State')
plt.ylabel('County Count')
plt.title('Distribution of Urban Influence Codes by State (2003 & 2013)')
plt.xticks(rotation=90)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
oil_gas_change_percentages = sub_county_df['Oil_Gas_Change'].value_counts(normalize=True) * 100

plt.figure(figsize=(8, 8))
plt.pie(oil_gas_change_percentages, labels=oil_gas_change_percentages.index, autopct='%1.1f%%', startangle=90)
plt.title('Percentage Distribution of Oil_Gas_Change Types')
plt.axis('equal')
plt.show()

plt.figure(figsize=(10, 6))
plt.bar(oil_gas_change_percentages.index, oil_gas_change_percentages.values)
plt.xlabel('Oil_Gas_Change Type')
plt.ylabel('Percentage')
plt.title('Percentage Distribution of Oil_Gas_Change Types')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
state_population_loss = sub_county_df.groupby('State')['Population_loss_2015_update'].sum().reset_index()
total_population_loss = state_population_loss['Population_loss_2015_update'].sum()
state_population_loss['Percentage'] = (state_population_loss['Population_loss_2015_update'] / total_population_loss) * 100

state_population_loss
```

```python
categories = ['Farming', 'Mining', 'Government', 'Recreation', 'Manufacturing', 'Retirement Destination']
category_columns = [f'Type_2015_{cat}_NO' if cat != 'Retirement Destination' else 'Retirement_Destination_2015_Update' for cat in categories]

if 'Retirement_Destination_2015_Update' in sub_county_df.columns:
    sub_county_df['Retirement_Destination_2015_Update'] = pd.to_numeric(sub_county_df['Retirement_Destination_2015_Update'], errors='coerce')
else:
    print("'Retirement_Destination_2015_Update' column not found in DataFrame.")


category_counts = sub_county_df[category_columns].sum()
total_counties = len(sub_county_df)

category_percentages = (category_counts / total_counties) * 100

category_percentages_df = pd.DataFrame({'Category': categories, 'Percentage': category_percentages})

category_percentages_df
```
